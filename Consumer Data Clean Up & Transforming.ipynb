# Loading A Sample of the Data Into Jupyter Notebook

```python
import pandas as pd
```

```python
raw_data = pd.read_excel("Consumer Spending 1.xlsx")
raw_data
```

```python
### data from a coworker
descript = pd.read_excel("Descriptions (1).xlsx")
descript = descript["Description"]
descript = descript[0:28]
descript
```

# Strip the word "category" in the Category column 

```python
raw_data["Category"] = raw_data["Category"].str.strip("category")
raw_data
```

```python
raw_data["Category"] = raw_data["Category"].str.strip("_")
raw_data
```

# Convert float to integer data type in End Age column to remove the ".0"

```python
raw_data["End Age"] = raw_data["End Age"].astype("int")
raw_data
```

# Set all elec, hobbies, clothing categories into Elec/Hobbies/Clothing because of high similarity/association

```python
raw_data.loc[raw_data["Category"] == "Elec", "Category"] = "Elec/Hobbies/Clothing"
raw_data.loc[raw_data["Category"] == "Hobbies", "Category"] = "Elec/Hobbies/Clothing"
raw_data.loc[raw_data["Category"] == "Clothing", "Category"] = "Elec/Hobbies/Clothing"
raw_data
```

# Set all furniture and building to Furniture/Building
# Set all misc and general to General/Misc

```python
raw_data.loc[raw_data["Category"] == "Furniture", "Category"] = "Furniture/Building"
raw_data.loc[raw_data["Category"] == "Building", "Category"] = "Furniture/Building"
raw_data
```

```python
raw_data.loc[raw_data["Category"] == "Misc", "Category"] = "General/Misc"
raw_data.loc[raw_data["Category"] == "General", "Category"] = "General/Misc"
raw_data
```

# Now reduce the number of duplicate rows

```python
raw_data[6:9] = raw_data[6:9].drop_duplicates()
raw_data[17:20] = raw_data[17:20].drop_duplicates()
raw_data[28:31] = raw_data[28:31].drop_duplicates()
raw_data[39:42] = raw_data[39:42].drop_duplicates()
raw_data[17:20] = raw_data[17:20].drop_duplicates()
raw_data[4:6] = raw_data[4:6].drop_duplicates()
raw_data[15:17] = raw_data[15:17].drop_duplicates()
raw_data[26:28] = raw_data[26:28].drop_duplicates()
raw_data[37:39] = raw_data[37:39].drop_duplicates()
raw_data[17:20] = raw_data[17:20].drop_duplicates()
raw_data[9:11] = raw_data[9:11].drop_duplicates()
raw_data[20:22] = raw_data[20:22].drop_duplicates()
raw_data[31:33] = raw_data[31:33].drop_duplicates()
raw_data[42:44] = raw_data[42:44].drop_duplicates()
raw_data
```

```python
### drop NAN rows
raw_data = raw_data.dropna()
raw_data
```

```python
### reset index
raw_data = raw_data.reset_index()
raw_data.head()
```

# Create columns Generation (Generation details), Generation and Color Order using the Year Born column

```python
raw_data["Generation (Generation details)"] = 1
raw_data["Generation"] = 1
raw_data["Color Order"] = 1
raw_data
```

```python
for i in list(range(len(raw_data))):
    if raw_data["Year Born"][i] == "1981-1997":
        raw_data["Generation"] = "Millenials"
    elif raw_data["Year Born"][i] == "1965-1980":
        raw_data["Generation"][i] = "Generation X"
    elif raw_data["Year Born"][i] == "1946-1964":
        raw_data["Generation"][i] ="Baby Boomers"
    else:
        raw_data["Generation"][i] = "Traditionalists"
raw_data
```

```python
raw_data["Generation (Generation details)"] = raw_data["Generation"]
raw_data["Color Order"] = raw_data["Generation"]
raw_data
```

# Join the description column given by coworker

```python
d = {'index': list(range(len(raw_data))), 'Description': descript}
description = pd.DataFrame(data=d)
```

```python
description
```

```python
raw_data = raw_data.merge(description, on="index", how="inner")
raw_data
```

```python
### remove the level_0 column
raw_data = raw_data.drop(columns="level_0").head()
raw_data
```

```python
# remove the index column
raw_data = raw_data.drop(columns="index").head()
raw_data
```
